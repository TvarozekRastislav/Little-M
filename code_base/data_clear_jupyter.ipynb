```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
plt.rcParams['figure.figsize'] = 20, 15
```

```python
#------------set-up---------------
data_in_name = "data_in.txt"
data_out_name = "data_out.csv"
separator = "|"
#---------------------------------
df = pd.read_csv('data_in.txt', sep=separator, header=None,skiprows=2)
df.columns = ["room","1","n","2","n","3","n"]
df.pop("n")
df = df.dropna(axis = 0)
df = df.astype({"3": int})
df
```

```python
plt.rcParams['figure.figsize'] = 30, 25
dic = {"size": 14}
matplotlib.rc('font', **dic)

sns.pairplot(df, hue='room', )
```
